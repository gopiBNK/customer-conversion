```python
import numpy as np
import pandas as pd
df=pd.read_csv("e-shop clothing 2008.csv",delimiter=';')
print(df)
df.isna().sum()
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
for x in df.columns:
    if df[x].dtype=="object":
        df[x]=encoder.fit_transform(df[x])
df
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
for x in df.columns:
    if df[x].dtype == "int":
        df[[x]]=scalar.fit_transform(df[[x]])
df

from sklearn.preprocessing import Normalizer
normalizer = Normalizer()
normalized_data= normalizer.fit_transform(df)
df = pd.DataFrame(normalized_data, columns=df.columns)
print(df)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

category_counts = df['page 1 (main category)'].value_counts()

# Bar chart
plt.figure(figsize=(10, 5))
sns.barplot(x=category_counts.index, y=category_counts.values, palette='viridis')
plt.title('Page Views by Main Category')
plt.xlabel('Main Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# Histogram for price
plt.figure(figsize=(8, 5))
sns.histplot(df['price'], bins=30, kde=True, color='blue')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

sns.pairplot(df[['price', 'price 2', 'order', 'session ID']], diag_kind='kde')
plt.show()

print("Average Session Duration:", df['session ID'].mean())

# Countplot for session lengths
plt.figure(figsize=(8, 5))
sns.histplot(df['session ID'], bins=30, kde=True, color='purple')
plt.title('Session Duration Distribution')
plt.xlabel('Session Duration')
plt.ylabel('Frequency')
plt.show()

# Assume bounce is where session has only 1 page viewed
df['bounced'] = df['page'].apply(lambda x: 1 if x == 1 else 0)

# Bounce rate
bounce_rate = df['bounced'].mean() * 100
print(f"Bounce Rate: {bounce_rate:.2f}%")

# Visualizing bounce rates
sns.barplot(x=['Bounced', 'Not Bounced'], y=[df['bounced'].sum(), len(df) - df['bounced'].sum()])
plt.title("Bounce Rate")
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 6))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Convert 'date' column to datetime format if needed
df['date'] = pd.to_datetime(df['year'].astype(str) + '-' + df['month'].astype(str) + '-' + df['day'].astype(str))

# Extract time-based features
df['hour'] = df['date'].dt.hour
df['day_of_week'] = df['date'].dt.day_name()

# Plot user activity by hour
plt.figure(figsize=(10, 5))
sns.countplot(x='hour', data=df, palette='coolwarm')
plt.title('User Activity by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Session Count')
plt.show()

```
